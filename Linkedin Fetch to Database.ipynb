```python
import pandas as pd
from googlesearch import search
from linkedin_api import Linkedin
import re

data=pd.read_csv('CompanyDatabase.csv') # Read the Master Company database
api = Linkedin("","") # Enter your Linkedin account details here
datadict = data.set_index('Name').T.to_dict('list')

CompanyNames=pd.read_csv('target_list_20201009 - list.csv') # Path to the CSV file of companies to find data for
CompanyNames = CompanyNames['Company Name'].tolist() # Choose columns where Company Names exist
CompanyInfo= [1] * len(CompanyNames)

for i in range(len(CompanyNames)):
    if CompanyNames[i] not in datadict:
        CompanyInfo[i] = 0

CompanyLinkedin=[]        
for i in range(len(CompanyNames)):
    if(CompanyInfo[i]==0):
        x = CompanyNames[i]
        x+=" Linkedin"
        j=0
        res = list(search(x, tld="com", num=10, stop=10, pause=2))
        while (j<len(res)) and ("linkedin.com/company/" not in res[j].lower()):
            j+=1
        
        flag=0
        parts = x.split()
        if(j>=len(res)):
            CompanyLinkedin.append("")
            continue
        print("{}.{}".format(i,res[j]))
        for part in parts:
            if res[j].find(part.lower()) > -1:
                flag=1
                CompanyLinkedin.append(res[j])
                break
        if flag==0:
            CompanyLinkedin.append("")
    else:
        CompanyLinkedin.append("")

CompanyLinkedinCode = []

for i in CompanyLinkedin:
    try:  
        CompanyLinkedinCode.append(re.search('company/(.*)',i).group(1))
    except:
        CompanyLinkedinCode.append('')

name = []
url = []
description = []
Locations = []
Employees = []
Domain = []
Industries = []
Specialities = []

for i in range(len(CompanyLinkedinCode)):
    if(CompanyLinkedinCode[i]!=''):
        try:
            name.append(CompanyNames[i])
            print(CompanyNames[i])
            Company = api.get_company(CompanyLinkedinCode[i])
            try:
                url.append(Company['url'])
            except:
                url.append('')
            try:
                description.append(Company['description'])
            except:
                description.append('')
            try:
                Domain.append(Company["companyPageUrl"])
            except:
                Domain.append('')
            try:
                Industries.append(Company["companyIndustries"])
            except:
                Industries.append('')
            try:
                Specialities.append(Company["specialities"])
            except:
                Specialities.append('')
            try:
                Employees.append(Company["staffCountRange"])
            except:
                Employees.append('')
        except:
            print('error') 
            url.append('')
            description.append('')
            Domain.append('')
            Industries.append('')
            Specialities.append('')
            Employees.append('')

CompanyInfodf = pd.DataFrame(list(zip(name,url,description,Employees,Domain,Industries,Specialities))) 
CompanyInfodf.to_csv('CompanyDatabase.csv', mode='a', header=False,index=False) #Updates the Master Company Database
```
